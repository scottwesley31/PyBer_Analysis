```python
# 5.1.7 Create Scatter Plots and Bubble Charts
import matplotlib.pyplot as plt
```

```python
x = [50, 42, 39, 34, 32, 48, 22, 17, 8, 13]
y = [17, 19, 21, 22, 24, 23, 34, 31, 43, 35]
```

```python
plt.scatter(x,y, c='green', marker='x', s=x)
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.title('Average Fare vs Total Rides')
plt.show()
```

```python
# 5.1.7 Create Pie Charts
import matplotlib.pyplot as plt
```

```python
categories = ['pants', 'shirts', 'outerwear', 'pajamas', 'underwear', 'shoes']
```

```python
revenue = [25, 15, 5, 20, 10, 25]
```

```python
plt.pie(revenue)
plt.show()
```

```python
plt.pie(revenue, labels=categories)
plt.show()
```

```python
# Include percentages of each slice using autopct=
plt.pie(revenue, labels=categories, autopct='%1.1f%%')
plt.show()
```

```python
# Assign which element in the list should be "exploded" out of the pie chart. The quantity of the number indicates how far
# out the slice will be from the whole.
explode = [0,.1,0,0,0,0]
```

```python
# Breakaway one of the pie slices to highlight it using explode=
plt.pie(revenue, labels=categories, autopct='%1.1f%%', explode=explode)
plt.show()
```
