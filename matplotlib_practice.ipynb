```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt(); can also be written this way:
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a legend.
plt.legend()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add a legend.
plt.legend()
```

```python
# Skill drill: Using the object-oriented approach, make the following changes to your line chart.
# Create the plot with ax.plt()
# Create the plot and add a label for the legend.
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label="Boston")

# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)

# Create a title.
ax.set_title("PyBer Fare by Month")

# Add a grid.
ax.grid()

# Add a legend.
ax.legend()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Create a list with y_axis values scaled up by 3x.
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
# Use y_axis_larger as marker size.
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Use list comprehension technique to multiply each data point in the y_axis list.
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create scatter plot using object-oriented interface.
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a bubble plot using object-oriented interface.
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Skill Drill 5.1.7; Bubble Chart
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color="skyblue", s = [i * 5 for i in y_axis], alpha=0.2, edgecolor='black', linewidth=2, label='Boston')
plt.gca().invert_yaxis()

# Create labels for the x and y axes.
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

# Set the y limit between 0 and 45.
ax.set_xlim(0, 50)

# Create a title.
ax.set_title("PyBer Fare by Month")

# Add a legend.
ax.legend()
```
